```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import pandas as pd
import pdfplumber
import os

# Configurar la ruta a Tesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Ruta del archivo PDF proporcionado (usando barras dobles)
pdf_path = r'C:\xampp\htdocs\php5\py\mnt\data\Convertir.pdf'
output_excel = r'C:\xampp\htdocs\php5\py\mnt\data\output_table.xlsx'

# Asegurarse de que la ruta de salida existe
output_dir = os.path.dirname(output_excel)
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Función para extraer imágenes de las páginas del PDF
def extract_images_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    images = []
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        images.append(img)
    return images

# Función para realizar OCR en una imagen y extraer texto
def extract_text_with_ocr(image):
    text = pytesseract.image_to_string(image)
    return text

# Procesar el PDF con pdfplumber para extraer tablas si están estructuradas
def extract_table_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        tables = []
        for page in pdf.pages:
            table = page.extract_table()
            if table:
                tables.append(table)
        return tables

# Extraer tablas usando pdfplumber (si es posible)
tables = extract_table_from_pdf(pdf_path)

# Si no se encontraron tablas, extraer las imágenes y realizar OCR
if not tables:
    print("No se encontraron tablas estructuradas. Intentando OCR...")
    images = extract_images_from_pdf(pdf_path)
    ocr_results = [extract_text_with_ocr(image) for image in images]
    
    # Imprimir los resultados del OCR
    for i, result in enumerate(ocr_results):
        print(f"Texto extraído de la página {i+1}:")
        print(result)
    
    # Guardar el resultado OCR en Excel como texto
    df = pd.DataFrame(ocr_results, columns=['Texto OCR'])
    df.to_excel(output_excel, index=False)
    print(f'OCR extraído y guardado en {output_excel}')
else:
    # Convertir las tablas a un DataFrame de pandas y luego a Excel
    print(f"Se encontraron {len(tables)} tabla(s) estructurada(s).")
    df = pd.DataFrame(tables[0])  # Suponiendo que solo queremos la primera tabla
    df.to_excel(output_excel, index=False)
    print(f'Tabla extraída y guardada en {output_excel}')
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import pandas as pd
import os

# Configurar la ruta a Tesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Ruta del archivo PDF proporcionado (usando barras dobles)
pdf_path = r'C:\xampp\htdocs\php5\py\mnt\data\Convertir.pdf'
output_excel = r'C:\xampp\htdocs\php5\py\mnt\data\tabla_convertida_ocr.xlsx'

# Asegurarse de que la ruta de salida existe
output_dir = os.path.dirname(output_excel)
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Función para extraer imágenes de las páginas del PDF
def extract_images_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    images = []
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        images.append(img)
    return images

# Función para realizar OCR en una imagen y extraer texto
def extract_text_with_ocr(image):
    text = pytesseract.image_to_string(image, config='--psm 6')
    return text

# Extraer imágenes de las páginas del PDF
images = extract_images_from_pdf(pdf_path)

# Aplicar OCR a cada imagen
ocr_results = [extract_text_with_ocr(image) for image in images]

# Imprimir los resultados del OCR
for i, result in enumerate(ocr_results):
    print(f"Texto extraído de la página {i+1}:")
    print(result)

# Guardar el resultado OCR en Excel como texto
df = pd.DataFrame(ocr_results, columns=['Texto OCR'])
df.to_excel(output_excel, index=False)
print(f'OCR extraído y guardado en {output_excel}')
```
