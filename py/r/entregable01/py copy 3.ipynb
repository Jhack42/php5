```python
import h2o
from h2o.estimators import H2OGradientBoostingEstimator, H2ORandomForestEstimator, H2OXGBoostEstimator
from h2o.frame import H2OFrame
h2o.init()

# Cargar el dataset
ruta_archivo = "C:/xampp_php5/htdocs/php5/py/r/entregable01/notificaciones_academicas.csv"
data = h2o.import_file(ruta_archivo)
```

```python
x_vars = ["Rol", "Actividad", "DiasAntes", "MinutosAntes", "HorasDespues", "TipoNotificacion"]
y_var = "Efectividad"
```

```python
gbm_model_efectividad = H2OGradientBoostingEstimator(ntrees=50, max_depth=5, learn_rate=0.1, seed=1234)
gbm_model_efectividad.train(x=x_vars, y=y_var, training_frame=data)
```

```python
print("Importancia de Variables para Efectividad (Necesidad de Recordatorios):")
gbm_model_efectividad.varimp_plot()
```
