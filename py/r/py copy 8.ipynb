```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
fahrenheit = np.array(range(-100, 201), dtype=float)  # Rango de temperaturas en Fahrenheit
celsius = (fahrenheit - 32) * 5.0 / 9.0  # Convertir a Celsius
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(fahrenheit, celsius)
plt.xlabel("Temperatura (°F)")
plt.ylabel("Temperatura (°C)")
plt.title("Datos de entrenamiento: Fahrenheit a Celsius")
plt.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])  # Capa densa con 1 neurona y una sola entrada
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')
```

```python
history = model.fit(fahrenheit, celsius, epochs=500, verbose=0)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'])
plt.xlabel("Épocas")
plt.ylabel("Pérdida (Error cuadrático medio)")
plt.title("Gráfico de pérdida durante el entrenamiento")
plt.show()
```

```python
fahrenheit_test = np.array([32, 100, 212, -40], dtype=float)
celsius_test_pred = model.predict(fahrenheit_test)
```

```python
for i, f in enumerate(fahrenheit_test):
    print(f"{f} °F -> {celsius_test_pred[i][0]:.2f} °C (esperado: {(f - 32) * 5.0 / 9.0:.2f} °C)")
```
